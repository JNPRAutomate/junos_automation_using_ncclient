### Execute JUNOS commands or JUNOS RPC

Open a NETCONF session with a Junos device.  
Add `device_params={'name':'junos'}` as an argument of the class connect, so you can even better manage Junos devices. 

```python
from ncclient import manager
vmx1=manager.connect(host="100.123.1.0", port="830", username="jcluser", password="Juniper!1",  timeout=60, hostkey_verify=False, device_params={'name':'junos'})
```

```python
vmx1.connected
```

Execute an rpc


```python
rpc = "<get-bgp-summary-information></get-bgp-summary-information>"
rpc_response = vmx1.rpc(rpc)
print(rpc_response)
```

Run a command and get output in text format 

```python
command = 'show bgp summary'
```

```python
command_output = vmx1.command(command=command, format='text')
print(command_output)
```

Run a command and get output in json format 

```python
command_output = vmx1.command(command=command, format='json')
print(command_output)
```

Run a command and get output in xml format 

```python
command_output = vmx1.command(command=command, format='xml')
print(command_output)
```

Use the `close_session` NETCONF operation to close the NETCONF session

```python
vmx1.close_session()
```
